```python
import pandas as pd
import xgboost as xgb
import numpy as np
from xgboost import plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import load_breast_cancer
```

```python
# Create arrays for the features and the target: X, y
dataset = load_breast_cancer()
X = dataset.data
y = dataset.target
```

```python
# Create the DMatrix: dataset_dmatrix
dataset_dmatrix = xgb.DMatrix(data=X, label=y)
```

```python
# Create the initial parameter dictionary for varying l2 strength: params
params={"objective":"binary:logistic","colsample_bytree":0.3,"learning_rate":0.2, "max_depth":4}
```

```python
# search for hyper-parameter for fine-tuning model, learning rate (learting rate/eta),
#gamma (min loss reduction to create new tree split), lambda (L2 reg on leaf weights), 
#alpha (L1 reg on leaf weights), max_depth (max depth per tree), subsample (%sample used per tree low:underfitting, high:overfitting)
#colsample_bytree (% features used per tree, large:all features overfitting, small:fraction of features kind of regularization)

reg_params = [1,10,100]


```

```python
# Create an empty list for storing accuracy as a function of l2 complexity
accuracy_reg = []
```

```python
# Iterate over reg_params and build one model per regularization parameter 
# Automated boosting round selection using early_stopping has been employed
for cl in reg_params:
    params["lambda"] = cl
    # Perform cross-validation: cv_results
    cv_results = xgb.cv(dtrain=dataset_dmatrix, params=params, nfold=10, early_stopping_rounds=10, num_boost_round=50, metrics="error", as_pandas=True, seed=123)
    
    # Append to final accuracy_reg
    accuracy_reg.append((1-cv_results["test-error-mean"]).iloc[-1])
```

```python
# Print cv_results
print(pd.DataFrame(list(zip(reg_params, accuracy_reg)), columns=["reg_params","accuracy_reg"]))
```

```python
# Grid search for hyperparameter tuning

gbm_param_grid = {'learning_rate':[0.01,0.1,0.5,0.9],
                 'n_estimators':[200],
                 'subsample':[0.3,0.5,0.9]}

# Instantiate the classifier

gbm = xgb.XGBClassifier()

# Perform grid search: grid_accuracy

grid_accuracy = GridSearchCV(estimator=gbm, param_grid=gbm_param_grid, scoring='accuracy', cv=4, verbose=1)

# Fit grid_accuracy to the data

grid_accuracy.fit(X,y)
```

```python
# Print the best parameters and highest accuracy

print("Best parameters found: ", grid_accuracy.best_params_)
print("Highest accuracy found: ", grid_accuracy.best_score_)
```

```python
# Random search for hyperparameter tuning

# Create the parameter for random search: gbm_param_random

gbm_param_random = {'learning_rate':np.arange(0.05,1.05,0.05),
                 'n_estimators':[200],
                 'subsample':np.arange(0.05,1.05,0.05)}

# Instantiate the classifier
gbm = xgb.XGBClassifier()

# Perform random search: randomized_accuracy
randomized_accuracy = RandomizedSearchCV(estimator=gbm, param_distributions=gbm_param_random, n_iter=25, scoring='accuracy', cv=4, verbose=1)

# Fit randomized_accuracy to the data
randomized_accuracy.fit(X,y)
```

```python
# Print the best parameters and highest accuracy
print("Best parameters found: ", randomized_accuracy.best_params_)
print("Highest accuracy found: ", randomized_accuracy.best_score_)
```

```python
# Create the training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

```python
# Instantiate the XGBClassifier: xg_cl
xg_cl = xgb.XGBClassifier(objective='binary:logistic',n_estimators=10,seed=123)
```

```python
# Fit the classifier to the training set
xg_cl.fit(X_train,y_train)
```

```python
# Predict the labels of the test set: preds
preds = xg_cl.predict(X_test)
```

```python
# Compute the accuracy: accuracy
accuracy = float(np.sum(preds==y_test))/y_test.shape[0]
```

```python
print('accuracy: %f'% (accuracy))
```
